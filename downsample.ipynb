# Параметры

```python
category_size = 500000
```

# Импорты

```python
import pandas as pd
from sklearn.utils import shuffle
```

# Считывание данных

```python
df_cell_phones = pd.read_json(r"dataset\Cell_Phones_and_Accessories\Cell_Phones_and_Accessories.json", lines=True, nrows=category_size)
df_clothing_jewelry = pd.read_json(r"dataset\Clothing_Shoes_and_Jewelry\Clothing_Shoes_and_Jewelry.json", lines=True, nrows=category_size)
df_electronics = pd.read_json(r"dataset\Electronics\Electronics.json", lines=True, nrows=category_size)
df_home_kitchen = pd.read_json(r"dataset\Home_and_Kitchen\Home_and_Kitchen.json", lines=True, nrows=category_size)
df_sports_outdoors = pd.read_json(r"dataset\Sports_and_Outdoors\Sports_and_Outdoors.json", lines=True, nrows=category_size)
df_toys_games = pd.read_json(r"dataset\Toys_and_Games\Toys_and_Games.json", lines=True, nrows=category_size)
df = pd.concat([df_cell_phones, df_clothing_jewelry, df_electronics, df_home_kitchen, df_sports_outdoors, df_toys_games], ignore_index=True)
df = shuffle(df).reset_index(drop=True)
```

```python
df['class'].value_counts()
```

```python
df_fake = df[df['class'] == 1]
df_true = df[df['class'] == 0]
df_fake_downsampled = df_fake.sample(df_true.shape[0])
```

```python
df = pd.concat([df_fake_downsampled, df_true])
df['class'].value_counts()
```

```python
df.to_csv('amazon_reviews_small.csv', index = False)
```
